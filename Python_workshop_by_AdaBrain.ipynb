## Exercise: Calculating Perfect Squares


The goal of this exercise is to prompt the user to enter a given number and find out whether it is a perfect square.


unnamed.gif

```
num = input("Enter a number to check is a perfect square: ") # String
num = abs(int(num))

is_perfect = False
i = 1

while i <= num**(0.5):
    if i*i == num:
        print(f'{num} is a perfect square!')
        is_perfect = True
    i += 1

if not is_perfect:
    print(f'{num} is not a perfect square!')
        
```

## The for Loop


The for loops are similar to while loops, but they have additional advantages, such as being able to iterate over strings and other objects.

**For "String"**

```
for s in "Ada":
    print(s)

word = "Words"
for w in word:
    print(w)
```

**For Range**

```
for i in range(9):
    print(i, end="-")
else:
    print(i+1)
```

## The continue Keyword

continue is another Python keyword designed for loops. When Python reaches the continue keyword, it stops the code and goes back to the beginning of the loop. 

```
for num in range(1,100):
  if num % 2 == 0:
    continue
  if num % 3 == 0:
    continue
  if num % 5 == 0:
    continue
  if num % 7 == 0:
    continue
  print(num)
```
